 ## Step 3. Data Preprocessing & Refinement
In this stage, we address the inconsistencies identified during the initial data exploration. Our primary goal is to align the datasets into a unified format to facilitate seamless merging and accurate time-series analysis.
 
 ### 3.1. Categorical Alignment and Filtering
In this step, we ensure that the categorical variables—specifically **Industry Names** and **District Names**—are perfectly aligned between the 2017 and 2018 datasets. Discrepancies in these categories can lead to biased growth or decline calculations.

**A. Initial Category Observation**
We first inspected the unique values for each categorical column. While the counts were similar, a deeper check was required to confirm if the actual labels matched across both years.

* Observations: * `Year_Month`, `City/County_Life_Name`, and `gender` showed consistent values.

    * Eup/myeon/dong_name and Industry_name required a set-difference check to identify year-specific entries.

```python
def print_unique_values(df):
    object_columns = df.columns[df.dtypes == 'object']
    for col in object_columns:
        print(f"Number of unique values in the '{col}' column: {df[col].nunique()}")
        print(sorted(df[col].unique()), '\n')

print_unique_values(jeju_reg_17_df)
```

```python
print_unique_values(jeju_reg_18_df)
```

**B. Identifying and Handling Exclusive Industry Values**

By comparing the `Industry_name` sets, we identified industries that appeared in only one of the two years.

* **Unique to 2017:** `Other gambling and betting businesses`

* **Unique to 2018:** `taxi transportation industry`

**Action Taken**: Since these industries do not have corresponding data in the opposing year, including them would skew the Year-over-Year (YoY) analysis. We explicitly removed these records to maintain a consistent baseline for comparison.

```python
for item in jeju_reg_17_df['Industry_name'].unique():
    if item not in jeju_reg_18_df['Industry_name'].unique():
        print(f'Unique values in 2017 dataset: {item}')
        
for item in jeju_reg_18_df['Industry_name'].unique():
    if item not in jeju_reg_17_df['Industry_name'].unique():
        print(f'Unique values in 2018 dataset: {item}')
```

```python
print(jeju_reg_17_df[jeju_reg_17_df['Industry_name'] == 'Other gambling and betting businesses'].shape)
print(jeju_reg_18_df[jeju_reg_18_df['Industry_name'] == 'taxi transportation industry'].shape)
```

```python
jeju_reg_17_df = jeju_reg_17_df[jeju_reg_17_df['Industry_name'] != 'Other gambling and betting businesses']
jeju_reg_18_df = jeju_reg_18_df[jeju_reg_18_df['Industry_name'] != 'taxi transportation industry']
```

**C. Verification of District Names (Eup/myeon/dong)**

We performed the same set-difference check for `Eup/myeon/dong_name` to ensure that all 43 administrative districts are named consistently and exist in both datasets. This ensures that our regional analysis is geographically accurate across the entire timeframe.

```python
for item in jeju_reg_17_df['Eup/myeon/dong_name'].unique():
    if item not in jeju_reg_18_df['Eup/myeon/dong_name'].unique():
        print(f'Unique values in 2017 dataset: {item}')
        
for item in jeju_reg_18_df['Eup/myeon/dong_name'].unique():
    if item not in jeju_reg_17_df['Eup/myeon/dong_name'].unique():
        print(f'Unique values in 2018 dataset: {item}')
```

### 3.2. Standardizing Date Formats
To enable cross-dataset analysis, we will:
1. Convert the string-based dates in the spending data into `datetime` objects.
2. Parse the numeric dates in the population dataset (`YYYYMMDD`) into the same `datetime` format.
This standardization is essential for accurate data merging and chronological visualization.

```python
# Convert Spending Data (Year_Month: String 'YYYY-MM-DD' -> Datetime)
jeju_reg_17_df['Year_Month'] = pd.to_datetime(jeju_reg_17_df['Year_Month'])
jeju_reg_18_df['Year_Month'] = pd.to_datetime(jeju_reg_18_df['Year_Month'])

# Convert Population Data (year_month_day: Integer 20170101 -> Datetime)
jeju_pop_df['year_month_day'] = pd.to_datetime(jeju_pop_df['year_month_day'].astype(str), format='%Y%m%d')

# Verify the results
print("--- Data Types After Conversion ---")
print(f"Spending 2017 Date Type: {jeju_reg_17_df['Year_Month'].dtype}")
print(f"Spending 2018 Date Type: {jeju_reg_18_df['Year_Month'].dtype}")
print(f"Population Date Type: {jeju_pop_df['year_month_day'].dtype}")

# Display the first few rows to confirm the visual format
display(jeju_reg_17_df[['Year_Month']].head(1))
display(jeju_reg_18_df[['Year_Month']].head(1))
display(jeju_pop_df[['year_month_day']].head(1))
```

### 3.3. Refining Date Resolution (String Slicing)
Upon inspecting the `Year_Month` column, we observed that while the data is stored as a `datetime` object in a `YYYY-MM-DD` format, the day values are consistently set to the 1st of each month (e.g., `2017-01-01`). Since our analysis focuses on monthly trends rather than daily fluctuations, maintaining the day suffix is unnecessary and may clutter visualizations.

**Action Taken:**
To streamline the dataset, we utilized the `.dt.strftime()` method to reformat the date objects into a cleaner `YYYY-MM` string format. This ensures the data reflects a monthly grain while remaining human-readable for reporting.

* **Method**: Applied `.dt.strftime('%Y-%m')` to the `Year_Month` column.

* **Objective**: To simplify the time-series axis for future plots and ensure the date grain matches our monthly analysis goals.

```python
jeju_reg_17_df['Year_Month'] = jeju_reg_17_df['Year_Month'].dt.strftime('%Y-%m')
jeju_reg_17_df.head()
```

```python
jeju_reg_18_df['Year_Month'] = jeju_reg_18_df['Year_Month'].dt.strftime('%Y-%m')
jeju_reg_18_df.head()
```

### 3.4. Data Integration (Concatenation)
With the preprocessing and categorical alignment complete, we now merge the separate 2017 and 2018 spending datasets into a single, unified DataFrame. This consolidated dataset allows us to perform seamless Year-over-Year (YoY) comparisons and identify long-term economic trends in Jeju.

**Action Taken:**
We utilized the `pd.concat()` function to vertically stack the two DataFrames. The resulting DataFrame is stored in a new variable called `jeju_reg_df`.

* **Method:** `pd.concat([jeju_reg_17_df, jeju_reg_18_df])`

* **Verification:** The combined DataFrame consists of **54,146 rows** and **7 columns**, confirming that all filtered records from both years have been successfully integrated.

```python
jeju_reg_df = pd.concat([jeju_reg_17_df, jeju_reg_18_df])

jeju_reg_df.shape
```

### 4.1. Analysis of Spending by Industry
In this section, we explore card usage patterns across different business sectors. We aim to identify which industries dominate the Jeju market in terms of total volume, popularity, and high-value consumption.

**A. Identifying Top Industries by Total Usage and Popularity**

We first aggregated the data by `Industry_name` to calculate the total usage amount and the number of users for each sector.

* **By Total Usage Amount**: High-volume industries such as **Korean restaurants, Supermarkets, and Gas stations** ranked at the top, indicating they are the primary drivers of the local economy.

* **By Number of Users**: When ranked by the count of unique users, **Chain convenience stores** climbed to the top, highlighting their high accessibility and frequent usage by both residents and tourists.

```python
groupby_sector = jeju_reg_df.groupby('Industry_name').sum(numeric_only=True)
groupby_sector.sort_values(by='Usage_amount', ascending=False).head(10)
```

```python
groupby_sector.sort_values(by='Number_of_users', ascending=False).head(10)
```

**B. Feature Engineering: Spending per User**

To understand the spending intensity of each customer, we created a new metric: `Spending_per_user`.

`Spending_per_user` = `Usage_amount` / `Number_of_users`

**Observation**: While convenience stores have the highest number of users, their Spending per user is relatively low (under 10,000 KRW), reflecting the nature of small, frequent transactions.

```python
groupby_sector['Spending_per_user'] = groupby_sector['Usage_amount'] / groupby_sector['Number_of_users']
groupby_sector.sort_values(by='Number_of_users', ascending=False).head(10)
```

**C. Data Refinement: Handling Outliers**

During the analysis of high-value spending, the **Bus transportation** sector appeared to have the highest per-user spending. However, a closer look revealed it had only **5 users** in total. This small sample size skews the average and does not represent a general trend.

**Action Taken:** We filtered out the `Bus transportation` sector to ensure the integrity of our high-value spending analysis.

```python
groupby_sector.sort_values(by='Spending_per_user', ascending=False).head(10)
```

```python
groupby_sector.sort_values(by='Number_of_users').head(10)
```

```python
jeju_reg_df = jeju_reg_df[jeju_reg_df['Industry_name'] != 'bus transportation']
```

**D. Final Insights: High-Value Consumption Sectors**

After refining the data, we re-evaluated the industries with the highest per-user spending.

* **Results:** **General entertainment bars, Hotel industries, and Other bar businesses** showed the highest spending per person.

* **Conclusion:** This indicates that while retail and dining (restaurants/convenience stores) drive the volume of transactions, the hospitality and nightlife sectors drive high-value individual consumption on the island.

```python
groupby_sector = jeju_reg_df.groupby('Industry_name').sum(numeric_only=True)
groupby_sector['Spending_per_user'] = groupby_sector['Usage_amount'] / groupby_sector['Number_of_users']

groupby_sector.sort_values(by='Spending_per_user', ascending=False).head(10)
```

### 4.3. High-Value Districts and Local Industry Activity

In this section, we investigate the relationship between geographic location and spending intensity. We identify districts with the highest **spending per capita** and analyze which specific industries drive the economy in these high-value areas. 

**A. Identifying Districts with Highest Spending per Capita**

To find the regions where individual consumers spend the most, we calculated the `Spending_per_user` for each administrative district.

* **Top 5 Districts:** Yerae-dong, Yeongcheon-dong, Yongdam 2-dong, Yeon-dong, and Ildo 1-dong.

* **Observation:** These districts represent the "premium" spending zones of Jeju, where the average transaction value per person is significantly higher than in other regions.

```python
groupby_reg['Spending_per_user'] = groupby_reg['Usage_amount'] / groupby_reg['Number_of_users']
groupby_reg.sort_values(by='Spending_per_user', ascending=False).iloc[:5]
```

```python
top5_region = groupby_reg.sort_values(by='Spending_per_user', ascending=False).iloc[:5].index
top5_region
```

**B. Top 5 Industries in High-Value Districts**

For the top 5 districts identified above, we extracted the most active industries based on total `Usage_amount`. This allows us to understand the unique economic character of each high-value zone.

* **Yerae-dong:** Dominated by the **Hotel industry and Golf course operations.**

* **Yongdam 2-dong:** Characterized by travel-related spending, including **Duty-free shops and Car rentals** (likely due to its proximity to the airport).

* **Ildo 1-dong:** Shows high activity in **Specialized retail**, such as Sports equipment and Seafood.

```python
groupby_reg_sec = jeju_reg_df.groupby(['Eup/myeon/dong_name', 'Industry_name']).sum(numeric_only=True).reset_index()

for reg in top5_region:
    reg_df = groupby_reg_sec[groupby_reg_sec['Eup/myeon/dong_name'] == reg]
    print(reg, reg_df.sort_values(by='Usage_amount', ascending=False).iloc[:5]['Industry_name'].tolist())
```
